# YOLO Custom Model Inference
---
## Load trained YOLO model

Load a custom-trained YOLOv8 detection model from local weights.

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

model = YOLO("runs/detect/circle_exp/weights/best.pt")
```

---
## Run inference on a validation image

Perform object detection on a single image from the validation dataset.

```python
results = model("test_img.jpg")
```

---
## Visualize detection results

Display bounding boxes and class labels predicted by the model.

```python
annotated = results[0].plot()

plt.figure(figsize=(6, 6))
plt.imshow(annotated)
plt.axis("off")
plt.show()
```
